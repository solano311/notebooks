```python
import pandas as pd # type: ignore
dados_ibov = pd.read_csv('^BVSP.csv')
dados_ibov.columns = dados_ibov.columns.str.replace(' ', '_').str.lower()
dados_ibov.set_index('date', inplace=True)
dados_ibov['return_close'] = dados_ibov.close.pct_change()
dados_ibov['return_adj_close'] = dados_ibov.adj_close.pct_change()
dados_ibov.head()
```

```python
dados_ibov.plot.line(y='close', style='-', color='r', title="Fechamento IBOV", xlabel='Data')
```

```python
dados_ibov.plot.hist(column='return_close', bins=10, title="Retorno Fechamnto IBOV", ylabel="Frequência",)
```

```python
((1 + dados_ibov.return_close.dropna()).cumprod() - 1).plot()
```
